#### Correlação das colunas categóricas por Cramer

Para avaliar a associação entre variáveis categóricas, utilizamos o coeficiente de associação de Cramer's V. Esta medida é uma versão ajustada do coeficiente qui-quadrado, que leva em consideração o tamanho da tabela de contingência. Valores próximos a 1 indicam uma associação forte entre as variáveis, enquanto valores próximos a 0 indicam uma associação fraca.

```python
def cramers_v(x, y):
    """Calculate Cramer's V statistic for categorical-categorical association."""
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Colunas Categoricas
categorical_cols = df.select_dtypes(include=['object']).columns

# Calculate Cramer's V for all pairs of categorical columns
cramer_v_matrix = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float)
for col1 in categorical_cols:
    for col2 in categorical_cols:
        cramer_v_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

print("Cramer's V Matrix:")
plt.figure(figsize=(10, 8))
sns.heatmap(cramer_v_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlação das Variáveis Categóricas (Cramer)')
plt.xlabel('Variáveis')
plt.ylabel('Variáveis')
plt.show()
```

### Testes de aderência

#### Distribuição Normal
Para verificar se as distribuições das variáveis numéricas seguem uma distribuição normal, realizamos o teste de normalidade de D'Agostino e Pearson. Este teste calcula uma estatística de teste e um valor p, onde um valor p abaixo de um determinado nível de significância (geralmente 0.05) indica que os dados não seguem uma distribuição normal.


```python
for col in df_numerical.columns:
    stat, p = normaltest(df_numerical[col])
    print(f'Coluna: {col}')
    print(f'Statistic: {stat}')
    print(f'p-value: {p}')
    if p > 0.05:
        print('A distribuição parece ser normal (não podemos rejeitar a hipótese nula)')
    else:
        print('A distribuição não parece ser normal (rejeitamos a hipótese nula)')
    print('\n')
```

Esses resultados indicam que, para as colunas numéricas listadas, a distribuição dos dados não parece seguir uma distribuição normal, com base nos testes de normalidade realizados.

O valor do teste estatístico é significativamente alto em todas as colunas, indicando que as distribuições dos dados divergem consideravelmente da normalidade.
Os valores de p (p-value) associados a cada teste são muito baixos, todos inferiores a 0.05, o que indica que há evidências estatísticas suficientes para rejeitar a hipótese nula de que os dados seguem uma distribuição normal.
No entanto, é importante observar que a coluna "scoreRC" possui um valor de p acima de 0.05, indicando que não há evidências estatísticas para rejeitar a hipótese nula nesse caso específico. Assim, para essa coluna, não podemos afirmar com confiança que os dados não seguem uma distribuição normal.

Em resumo, para a maioria das colunas numéricas listadas, os dados não parecem seguir uma distribuição normal. Isso pode ter implicações importantes ao realizar análises estatísticas que assumem normalidade nos dados. Seria prudente considerar métodos estatísticos alternativos que não exijam essa suposição ou explorar transformações nos dados para torná-los mais aproximados de uma distribuição normal, se apropriado para o seu contexto de análise.

#### Teste Qui Quadrado
Utilizamos o teste do qui-quadrado de independência para avaliar se duas variáveis categóricas estão associadas de forma independente ou se há uma associação significativa entre elas. O teste compara as frequências observadas em uma tabela de contingência com as frequências esperadas sob a hipótese nula de independência.


**- Verificar valor esperado, e se não zerou**

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

for col in df_cat.columns:
    observed_values = pd.crosstab(index=df_cat[col], columns='count').values.flatten()
    expected_values = np.full_like(observed_values, fill_value=observed_values.mean())
    chi2, p, dof, expected = chi2_contingency(observed_values.reshape(1, -1))

    print(f'Coluna: {col}')
    print(f'Chi-square statistic: {chi2}')
    print(f'p-value: {p}')
    print('Valores Esperados:')
    print(expected.reshape((len(observed_values), -1)))

    alpha = 0.05
    if p > alpha:
        print('A distribuição parece ser consistente com a esperada (não podemos rejeitar a hipótese nula)')
    else:
        print('A distribuição não parece ser consistente com a esperada (rejeitamos a hipótese nula)')
    print('\n')
```


Isso significa que, para todas as colunas categóricas que foram testadas, o teste do qui-quadrado não encontrou evidências suficientes para rejeitar a hipótese nula de que a distribuição observada das categorias é consistente com a distribuição esperada.

O qui-quadrado estatístico foi calculado como 0 para todas as colunas. O p-valor associado a cada teste foi calculado como 1.0. Um p-valor de 1.0 significa que não há evidências estatísticas para rejeitar a hipótese nula. Portanto, em todos esses casos, não há diferença estatisticamente significativa entre a distribuição observada das categorias e a distribuição esperada.

Em resumo, os resultados indicam que, com base nos dados disponíveis, não há razão para suspeitar que as distribuições das categorias em suas colunas categóricas sejam diferentes das distribuições esperadas.

## 2.3 Seleção de Variáveis

A estratégia de seleção de atributos aqui implementada terá como objetivo remover variáveis que não diminuem a incerteza sobre a nossa variável dependente. Para isso, para cada atributo real será criado um equivalente com a sua ordem aleatorizada, de modo que qualquer associação com a variável dependente será anulada. Após isso, atributos com informação mútua superior ao que foi gerado aleatoriamente será mantido.

```python
import numpy as np
from sklearn.feature_selection import mutual_info_classif

np.random.seed(42)

def select_important_features(X, y, features_to_shuffle=None):
    """
    Selects important features based on mutual information, comparing the
    mutual information of the original features with that of randomized versions.
    """
    # If no specific features to shuffle are given, shuffle all columns
    if features_to_shuffle is None:
        features_to_shuffle = list(X.columns)

    # Copy the original DataFrame to add randomized features
    X_ = X.copy()

    # List to hold the names of the randomized features
    mocked_features = []

    # Creating randomized features
    for feature in features_to_shuffle:
        X_[f"{feature}_randomized"] = np.random.choice(
            X[feature],
            size=len(X),
            replace=False
        )
        mocked_features.append(f"{feature}_randomized")

    # Calculating mutual information
    mut_info = mutual_info_classif(X_.values, y)

    # Finding the maximum mutual information value among the randomized features
    max_mocked_values = max(mut_info[len(features_to_shuffle):])

    # Selecting variables with mutual information greater than the max from randomized
    selected_variables = []
    for idx, feature in enumerate(features_to_shuffle):
        if mut_info[idx] > max_mocked_values:
            selected_variables.append(feature)

    return selected_variables
```

```python
selected_variables = select_important_features(X_transformed, y)
X_selected = X_transformed[selected_variables]
print('Variáveis selecionadas:')
print(selected_variables)
```

O t-SNE (t-distributed Stochastic Neighbor Embedding) é uma técnica de redução de dimensionalidade que é comumente usada para visualizar dados de alta dimensão em um espaço bidimensional ou tridimensional. Ele funciona mapeando os pontos de dados de alta dimensão para um espaço de menor dimensão, preservando ao máximo possível as distâncias entre os pontos originais. Isso permite visualizar padrões e agrupamentos nos dados que podem não ser evidentes na dimensionalidade original.

O algoritmo t-SNE opera em duas etapas principais:

1. **Construção de Probabilidades de Similaridade:** O t-SNE calcula probabilidades condicionais que representam as similaridades entre os pontos no espaço de alta dimensão. Ele usa uma distribuição t de Student para medir a similaridade entre pares de pontos, atribuindo maiores probabilidades a pares de pontos que são próximos em alta dimensão e menores probabilidades a pares de pontos que são distantes.
2. **Otimização da Embedding:** Em seguida, o algoritmo ajusta uma distribuição de probabilidade similar nos dados no espaço de baixa dimensão. Ele minimiza a divergência de Kullback-Leibler entre as distribuições de probabilidade de alta e baixa dimensão, ajustando a posição dos pontos no espaço de baixa dimensão de forma a refletir as similaridades entre os pontos no espaço de alta dimensão.

## 2.6 Analise de Agrupamentos: KMeans e DBSCAN

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score

# Definindo uma lista de parâmetros para KMeans e DBSCAN
kmeans_params = {
    'n_clusters': list(range(3, 21)),
    'init': ['k-means++', 'random'],
    'max_iter': [100, 150, 200, 250, 300],
}

dbscan_params = {
    'eps': [0.1, 0.5, 1.0],
    'min_samples': [3, 5, 7]
}

# Lista para armazenar os resultados
results = []

# Loop sobre os parâmetros do KMeans
for n_clusters in kmeans_params['n_clusters']:
    for init in kmeans_params['init']:
        for max_iter in kmeans_params['max_iter']:
            # Criando o pipeline do KMeans com os parâmetros atuais
            kmeans_pipeline = Pipeline([
                ("kmeans", KMeans(n_clusters=n_clusters, init=init, max_iter=max_iter, n_init=10))
            ])

            # Ajustando o modelo aos dados de treinamento
            kmeans_pipeline.fit(X_train)

            # Obtendo os rótulos
            kmeans_labels = kmeans_pipeline.named_steps['kmeans'].labels_

            # Calculando o silhouette score
            silhouette = silhouette_score(X_train, kmeans_labels)

            # Armazenando os resultados
            results.append({
                'algorithm': 'KMeans',
                'n_clusters': n_clusters,
                'init': init,
                'max_iter': max_iter,
                'silhouette_score': silhouette
            })

# Loop sobre os parâmetros do DBSCAN
for eps in dbscan_params['eps']:
    for min_samples in dbscan_params['min_samples']:
        # Criando o pipeline do DBSCAN com os parâmetros atuais
        dbscan_pipeline = Pipeline([
            ("dbscan", DBSCAN(eps=eps, min_samples=min_samples))
        ])

        # Ajustando o modelo aos dados de treinamento
        dbscan_pipeline.fit(X_train)

        # Obtendo os rótulos
        dbscan_labels = dbscan_pipeline.named_steps['dbscan'].labels_

        # Verificando se há pelo menos dois clusters distintos
        num_unique_labels = len(np.unique(dbscan_labels))
        if num_unique_labels > 1:
            # Calculando o silhouette score
            silhouette = silhouette_score(X_train, dbscan_labels)

            # Armazenando os resultados
            results.append({
                'algorithm': 'DBSCAN',
                'eps': eps,
                'min_samples': min_samples,
                'silhouette_score': silhouette
            })

# Convertendo os resultados para um DataFrame do pandas
results_df = pd.DataFrame(results)

# Encontrando o modelo com o melhor silhouette score
best_model = results_df.loc[results_df['silhouette_score'].idxmax()]

print("Melhor modelo:")
best_model
```

Este código realiza uma busca exaustiva de hiperparâmetros para os algoritmos de clusterização KMeans e DBSCAN e avalia a qualidade dos clusters gerados usando o índice de silhouette.


1. **Importações**: Importa as bibliotecas necessárias, incluindo numpy, pandas e os algoritmos de clusterização KMeans e DBSCAN do scikit-learn, bem como as métricas de avaliação, como o índice de silhouette.

2. **Definição de Parâmetros**: Define os parâmetros a serem testados para o KMeans e DBSCAN.

3. **Inicialização de Resultados**: Inicializa uma lista vazia `results` para armazenar os resultados de cada configuração de hiperparâmetros testada.

4. **Loop KMeans**: Itera sobre os parâmetros do KMeans. Para cada combinação, cria um pipeline do scikit-learn onde o modelo KMeans é ajustado aos dados de treinamento e o índice de silhouette é calculado.

5. **Loop DBSCAN**: Itera sobre os parâmetros do DBSCAN. Para cada combinação, cria um pipeline do scikit-learn onde o modelo DBSCAN é ajustado aos dados de treinamento. Antes de calcular o índice de silhouette, verifica-se se o DBSCAN encontrou pelo menos dois clusters distintos.

6. **Conversão para DataFrame**: Converte os resultados para um DataFrame do pandas chamado `results_df`.

7. **Identificação do Melhor Modelo**: Identifica o modelo com o melhor índice de silhouette.
